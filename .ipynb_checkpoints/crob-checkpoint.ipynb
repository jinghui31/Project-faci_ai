```python
import cv2
img = cv2.imread('S__43925508.jpg')
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),0)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
cut_img = img[y+10:y+h+10,x:x+w]
cv2.imwrite('S__43925508-1.jpg', cut_img)
```

```python
import cv2
img = cv2.imread("raw/raspberry.jpg")

#可以偵測，但是會有誤差，誤差會取1/4人臉
#face_cascade = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
#容易有誤差
#face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt_tree.xml')
#可以偵測到整張臉，若人臉有點歪，會有誤差，取2/3人臉
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
#face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray,scaleFactor=1.16,minNeighbors=5,minSize=(25,25),flags=0)

for (x,y,w,h) in faces:
    roi_gray = gray[ y : y + h, x : x + w]
    roi_color = img[ y : y + h, x : x + w]

cut_img = img[y : y + h, x : x + w]
cv2.imwrite('a21.jpg', cut_img)
```
