```python
from os import listdir
from os.path import join
from PIL import Image, ImageChops
import numpy as np
import cv2
```

```python
path = "raw"
path1 = "crob"
path2 = "resize"
path3 = "schannel"
path4 = "acne"
```

```python
files = listdir(path)
files.sort()

for f in files:
    file_path = join(path, f)
    img = cv2.imread(file_path)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.16, minNeighbors = 5, minSize = (25,25), flags = 0)

    for (x, y, w, h) in faces:
        roi_gray = gray[ y : y + h, x : x + w]
        roi_color = img[ y : y + h, x : x + w]
    
    file_path1 = join(path1, f)
    cut_img = img[y : y + h, x : x + w] 
    cv2.imwrite(file_path1, cut_img)
    
    #file_path2 = join(path2, f)
    #resize_img = Image.open(file_path1)
    #resize_img = resize_img.resize((300, 300), Image.ANTIALIAS)
    #resize_img.save(file_path2)
```

```python
files_re = listdir(path2)
files_re.sort()

for x in files_re:
    file_path = join(path2, x)
    hsv_img = cv2.imread(file_path)
    hsv_img = cv2.cvtColor(hsv_img, cv2.COLOR_BGR2HSV)
    
    file_path3 = join(path3, x)
    cv2.imwrite(file_path3, hsv_img[:,:,1])
    
    s_img = np.array(Image.open(file_path3))
    rows, cols = s_img.shape
    for i in range(rows):
        for j in range(cols):
            if (s_img[i , j] < 64):
                s_img[i , j] = 0
            elif (s_img[i , j] < 128):
                s_img[i , j] = 1
            elif (s_img[i , j] < 192):
                s_img[i , j] = 2
            else:
                s_img[i , j] = 3
    
    print(x, float(round(np.sum(s_img) / (rows * cols), 2)))
```

```python
files_acne = listdir(path2)
files_acne.sort()

for z in files_acne:
    file_path = join(path2, z)
    img = Image.open(file_path)
    width = img.size[0]
    height = img.size[1]
    #
    leftCheek = ( width / 6, height / 2, width / 3, height / 3 * 2)
    cropped_img = img.crop(leftCheek)
    #
    rightCheek = ( width / 3 * 2, height / 2, width / 6 * 5, height / 3 * 2)
    cropped_img2 = img.crop(rightCheek)
    #預留圖片
    toImage = Image.new('RGB', (100, 50))
    #將兩張圖片貼上
    toImage.paste(cropped_img, (0, 0))
    toImage.paste(cropped_img2, (50, 0))
    file_path4 = join(path4, z)
    toImage.save(file_path4)
```
