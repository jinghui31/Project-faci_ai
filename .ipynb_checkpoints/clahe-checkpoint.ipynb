```python
from os import listdir
from os.path import join
import cv2
from matplotlib import pyplot as plt

files = listdir('cheekForKNN_hsv/good/')
files.sort()

for f in files:
    file_path = join('acne', f)
    img = cv2.imread(file_path)

    # imgHSV
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # 
    img_v = img_hsv[:, :, 2]
    
    # V
    equal_img_v = cv2.equalizeHist(img_v)
    
    # VHSV
    img_hsv[:, :, 2] = equal_img_v

    # OpenCVimshowBGRHSVBGR
    equal_img = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)
    # 
    plt.hist(img.ravel(), 256, [0, 255])
    plt.hist(equal_img.ravel(), 256, [0, 255])
    plt.show()

    file_path1 = join('hsv', f)
    cv2.imwrite(file_path1, equal_img)
```
